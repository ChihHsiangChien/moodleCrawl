```python
# 取得目前開啟的測驗連結與名稱
id = 1735
#https://weteach.edu.tw/course/view.php?id=1735
courseurl = 'http://weteach.edu.tw/course/view.php?'
quiznameList=[]
quizurlList=[]
for i in range(7):
    url = courseurl + 'id='+str(id) +'&section='+str(i+1)
    driver.get(url)
    #取得各section原始碼
    htmltext = driver.page_source

    soup=BeautifulSoup(htmltext,"html.parser")
    
    #搜尋所有含quiz的連結
    links = soup.find_all('a',{'href':re.compile('quiz')})
    for link in links:

        #取得href屬性前先檢查是否存在
        if 'href' in link.attrs:
            href = link['href']
            print(link.text)
            print(href)
            quiznameList.append(link.text.replace(" 測驗卷", ""))
            quizurlList.append(href)
        else:
            href = None
```

```python
# 取得成績報表
scoreReportUrl = "https://weteach.edu.tw/grade/report/grader/index.php?id=1735"
driver.get(scoreReportUrl)
htmltext = driver.page_source

soup=BeautifulSoup(htmltext,"html.parser")
table = soup.find('table', {'class': 'gradereport-grader-table'})
#columns = [th.text.replace('\n', '') for th in table.find('tr').find_all('th')]

```

```python
# 從學生報表中取得成績標題與學生成績
# 學生成績標題
testTitleTrs = table.find_all('tr')[2]
titles = [title.text for title in testTitleTrs.find_all('th')[4:]] 
titles = ["Name"] + titles

# 學生成績
trs = table.find_all('tr')[3:]
names = list()
allscores = list()
for tr in trs:
    studentName =  tr.find_all('th')[0].text
    #names.append(studentName)
    scores = [studentName] + [td.text for td in tr.find_all('td')[4:]]
    allscores.append(scores)
#print(titles)
#print(names)

# 存入df
df = pd.DataFrame(allscores, columns =titles) 
df.set_index('Name', inplace = True)
df
```

```python
def deleteRecords(quizId, passingScore):
    #組合出該次考試成績報表的url
    url1 = "https://weteach.edu.tw/mod/quiz/report.php?id="
    url2 = "&mode=overview&attempts=enrolled_with&onlygraded&onlyregraded=0&slotmarks=1&tsort=sumgrades&tdir=4"
    url = url1 + quizId + url2    

    driver.get(url)
    
    #有多少人已經獲得成績
    quizattemptcounts = driver.find_element(By.CLASS_NAME, "quizattemptcounts").text
    quizattemptcounts = int(quizattemptcounts.replace("作答：", ""))


    #成績報表裡的欄位
    str1 = "mod-quiz-report-overview-report_r"

    #尋找低於passingScore的紀錄，並click。用clickNum統計有多少人低於passingScore
    clickNum = 0
    for i in range(quizattemptcounts):
        scoreId = str1 + str(i) + "_c10"
        #跳過作答中的紀錄
        if driver.find_element(By.ID, scoreId).text == "-":
            continue
        #成績轉成float
        ele = float(driver.find_element(By.ID, scoreId).text)
        #成績低於passingScore，則按下最前方的核取方塊，並將clickNum+1
        if ele < passingScore :
            print(ele)
            inputId =  str1 + str(i) + "_c0"
            inputEle = driver.find_element(By.ID, inputId)
            inputEle.click()
            clickNum += 1
    #如果沒有人被選取，則跳過此次紀錄
    if clickNum == 0 :
        return
    btnSubmit = driver.find_element(By.ID, "deleteattemptsbutton")
    btnSubmit.click()

    #刪除紀錄的[是】按鈕在按下btn後才會出現在另一個div裡頭。故先找到div再按鈕
    myDiv = driver.find_element(By.CLASS_NAME, "confirmation-dialogue")
    btnSubmit2 = myDiv.find_element(By.CLASS_NAME, "btn-primary")
    btnSubmit2.click()
    
```
